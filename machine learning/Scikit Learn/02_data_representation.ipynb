# Data representation

> Data as table
* 2D grid
* rows of the matrix are refer as samples.
* columns of the matrix are refer as features.

```python
from sklearn.datasets import load_iris
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

> Data as Feature matrix
* stores variable name `X` have 2D shape [n-sample,n_features]

> Data as target array
* also called label
* denoted by y
* usually have 1D having length n_samples
* generally contained in Numpy array or pandas series
* may contain continuous numerical or discrete values.
