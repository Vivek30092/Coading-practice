># Random numbers in numpy
* random no. means something that can not be predicted logically.
* Random number generated through a generation algorithm are called `pseudo random`.
* if there is a program to generate random number it can be predicted, thus it's not truly random.

`random` module is used to work with random number which is offered by numPy

```python
import numpy as np
from numpy import random

print(random.randint(100))   # random int
print(random.rand())        # random float
```

> create random 1D array

```python
print(random.randint(100,size=5))   # int array
print(random.rand(5))               # float array
```

> create random 2D array

```python
print(random.randint(100,size=(3,5)))   # int array
print(random.rand(3,4))                 # float array
```

### Generate Random Number from ARray

* The `choice()` method allows us to generate a random value based on an array of values 
*  it takes an array as a parameter and randomly returns one of the values.

```python
print(random.choice([1,4,6,7,14,30,87,43,23,2]))
```

we can give `size` to specify the shape of array `choice()` method and it return an array of values.

```python
print(random.choice([1,3,5442,21,42,5,35,2,3],size=(2,4)))
```

># Random Data Distribution
* data distribution is a list of all possible values, and how often each value occurs, those lists are important when working with statistics and data science.
* random module offer methods that returns randomly generated data distributions

`Probability Density Function`: A function that describes a continuous probability. i.e. probability of all values in an array.

* random numbers can be generated based on defined probabilities using the `choice()` method of the `random` module.
* the `Choice()` method allows us to specify the probability for each value.
* sum of all probability numbers should be `1`


```python
# Generate a 1-D array containing 100 values, where each value has to be 3, 5, 7 or 9.

# The probability for the value to be 3 is set to be 0.1

# The probability for the value to be 5 is set to be 0.3

# The probability for the value to be 7 is set to be 0.6

# The probability for the value to be 9 is set to be 0

#creating 1D array
print(random.choice([1,2,4,5,8,14,30,45,32],p=[0.0,0.1,0.1,0.1,0.1,0.2,0.3,0.1,0.0],size=(10)))

#creating 2D array
print(random.choice([1,2,4,5,8,14,30,45,32],p=[0.0,0.1,0.1,0.1,0.1,0.2,0.3,0.1,0.0],size=(2,5)))
```

```python
# creating 3D array
print(random.choice([1,2,4,5,8,14,30,45,32],p=[0.0,0.1,0.1,0.1,0.1,0.2,0.3,0.1,0.0],size=(2,5,3)))
```

># random Permutation

* permutation refers to an arrangement of elements eg [3,2,1] is a permutation of [1,2,3] and vice-versa.
* numpy random module provides two methods for this: `shuffle()` and ` permutation()`

### Shuffling arrays
changing arrangement of elements in-place. i.e in the array itself

```python
arr = np.array([1,2,3,4,5])
random.shuffle(arr)
print(arr)
```

`shuffle()` method makes changes to the original array.

### generating permutation of arrays
* generate a random permutation of elements of array
* `random()` method returns a re-arranged array(and leaves the original array un-changed).

```python
print(random.permutation(arr))
print(arr)
```

># Normal (Gaussian) Distribution
* one of the most imp distribution
* also called the Gaussian Distribution after the German mathematician `Carl Friedrich Gauss`
* fits the probability distribution of many events eg IQ Scores.
* use the `random.normal()` method to get a normal data distribution
* uses three parameter
- `loc` = (Mean) where the peak of he bell exists
- `scale` = (Standard Deviation) how flat the graph distribution should be
- `size` = the shape of the returned array.

```python
# generate a random distribution fo size 2X3
print(random.normal(size=(2,3)))
```

```python
# random normal distribution of size 2X3 with mean at 1 and standard deviation of 2:
print(random.normal(loc=1,scale=2, size=(2,3)))
```

### Visualization of normal distribution

`Note`: The curve of a Normal Distribution is also known as the `Bell Curve` because of the bell-shaped curve.

```python
sns.displot(random.normal(size=100),kind="kde")
plt.show()
```

># Binomial Distribution
* it's `discrete distribution`.
* describes the outcome of binary scenarios, eg. toss of a coin, it will either be head or tails.
* it has thee parameters:
- `n` = number of trials
- `p` = probability of occurence of each trial(eg.for toss of a coin 0.5 each).
- `size` = the shape of the returned array.

`Discrete Distribution`:The distribution is defined at separate set of events, e.g. a coin toss's result is discrete as it can be only head or tails whereas height of people is continuous as it can be 170, 170.1, 170.11 and so on.

```python
# 10 trials for coin toss generate 10 data points:

print(random.binomial(n=10,p=0.5,size=10))
```

### visualization of binomial distribution


```python
sns.displot(random.binomial(n=10,p=0.5,size=1000))
```

### Difference Between Normal and Binomial Distribution

The main difference is that normal distribution is continous whereas binomial is discrete, but if there are enough data points it will be quite similar to normal distribution with certain loc and scale.

```python
data = {
  "normal": random.normal(loc=50, scale=5, size=1000),
  "binomial": random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```
