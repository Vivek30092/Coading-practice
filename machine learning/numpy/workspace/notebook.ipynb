```python
# dtype as an argument
float32_array = np.array([1.25,52.25,52.22],dtype = np.float32)
float32_array.dtype
```

```python
# type conversion
bool_array = np.array([[True,False],[False,True]],dtype=np.bool_)
bool_array.astype(np.int32)
```

# type coercion

```python
# just becasue of one float value python has change data type of array into float64
print(np.array([0,23,432.21]).dtype)

# adding an int to an array of bool will change all boolean in to int
print(np.array([True,False,32]).dtype)
```
